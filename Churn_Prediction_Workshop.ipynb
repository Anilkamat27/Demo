# Meet with Stakeholders
Data Scientists/Applied Scientists rarely ever work in isolation.
Some great questions to ask your stakeholders would be:


1. **How Will the Predictions Be Used?**
   - Understanding the purpose of the churn predictions is essential. Are they intended for proactive customer retention, targeted marketing, or resource allocation? Knowing the end goal will guide model development and evaluation.

2. **Data Quality and Missing Values:**
   - **Can MonthlyCharges Have Missing Values in the Prediction Datasets? Why Might That Be? How Is the Data Generated?**
     - Investigate whether missing values exist in the `MonthlyCharges` feature. Understanding the reasons behind missing data can impact model performance.
     - Possible reasons for missing values:
       - **Data Collection Issues**: Was the data collected from various sources, leading to inconsistencies or gaps?
       - **Customer Behavior**: Some customers may not have monthly charges due to specific plans or promotions.
       - **Technical Errors**: Data extraction or recording errors might result in missing values.
     - Discuss strategies for handling missing data (imputation, removal, or treating them as a separate category).

3. **Additional Data:**
   - **Do You Have Demographic Information?**
     - Demographics (age, gender, location) can impact churn behavior.
   - **External Data:**
     - Explore external data (e.g., economic indicators, social media trends, competitors prices, etc.).

4. **Business Metrics and KPIs:**
   - **Which Metrics Define Churn?**
     - Clarify how churn is defined (e.g., cancellation, non-renewal, inactivity, downgrading packages).
   - **Other Relevant KPIs:**
     - Consider customer lifetime value (CLV), customer acquisition cost (CAC), and retention rate.
   

5. **Feature Importance:**
   - **Which Features Are Most Relevant for Churn Prediction?**
     - Prioritize features that strongly influence churn.
   - **Domain Knowledge:**
     - Leverage stakeholder expertise to identify critical features.

6. **Model Interpretability:**
   - **How Important Is Model Explainability?**
     - Some stakeholders may prioritize interpretability over model complexity.
   - **Business Rules:**
     - Discuss any business rules that override model predictions.

7. **Actionability:**
   - **What Actions Will Be Taken Based on Predictions?**
     - Ensure that predictions lead to actionable steps (e.g., personalized offers, retention campaigns).
   - **Cost of False Positives and Negatives:**
     - Understand the impact of misclassifying churners and non-churners.

8. **Validation and Deployment:**
   - **How Will the Model Be Validated?**
     - Discuss holdout datasets, cross-validation, and evaluation metrics.
   - **Integration with Existing Systems:**
     - Plan for model deployment and integration into operational workflows.

9. **Prediction Frequency:**
   - **At What Frequency Do You Want to Do the Predictions?**
     - Determine the prediction interval (e.g., daily, weekly, monthly) based on business needs.
     - Consider operational feasibility, computational resources, and the desired lead time for taking action based on predictions
      


**Remember, effective collaboration with stakeholders ensures that the churn prediction model aligns with business objectives and provides actionable insights. 📊🔍🤝**

### Format the data appropriately


```python
# we've already changed the dtype of Total Charges
# convert churn to binary numeric variable
df['Churn_onehotenc'] = np.where(df.Churn == 'Yes',1,0)
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, max(df['tenure']), 12)]

df['tenure_group'] = pd.cut(df.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
# Converting all categorical variables into dummy variables
variables_for_model = ['SeniorCitizen', 'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling','PaymentMethod', 'tenure_group', 'Churn', 'fraction_monthly_charge', 'Churn_onehotenc','MonthlyCharges', 'TotalCharges']
```

```python
df_with_dummies = pd.get_dummies(df[variables_for_model], dtype=np.uint8, drop_first=True)
```

### Feature Engineering

You can be creative here:
1. Handsoff People - People who have 1-2 year contracts and have paperless billing enabled, and are making automatic payments. These are people who value convenience a lot and would rather spend their time in other ways as opposed to hunting for better deals from other providers.
2. Streamers - People who like streaming movies/videos and have been with the company for a long enough time would probably not churn. The fact that they have been with the compaby long enough means they are happy with the streaming service and would probably not churn.

```python
handsoff_customers = df.loc[(df['Contract']!='Month-to-month') & (df['PaperlessBilling']=='Yes') & (df['PaymentMethod'].str.contains("automatic"))]
```

```python
df['handsoff_customers'] = df.apply(lambda s: 1 if ((s['Contract']!='Month-to-month')
                                                  & (s['PaperlessBilling']=='Yes')
                                                  & ("automatic" in s['PaymentMethod'] ))
                                                  else 0, axis=1)

df_with_dummies['handsoff_customers'] = df.apply(lambda s: 1 if ((s['Contract']!='Month-to-month')
                                                  & (s['PaperlessBilling']=='Yes')
                                                  & ("automatic" in s['PaymentMethod'] ))
                                                  else 0, axis=1)
```

```python
# let's see if there's a relation between handsoff customers and churn:
pd.crosstab(df['handsoff_customers'], df['Churn'], normalize="index")
```

**Insight:** Only 7% of the 'Hands Off' customers churn whereas ~30% of non-Hands-Off customers churn.

```python
# verify with chi square test
stat, p, dof, expected = chi2_contingency(pd.crosstab(df['handsoff_customers'], df['Churn']))
if p<0.05:
  print (f"Since p-value ({p}) is less than 0.05, we reject the null hypothesis that Churn is independent of whether or not a customer is Hands-Off")
```

### What is Precision and Recall?
image.png

Image Source: [Link](https://twitter.com/akshay_pachaar/status/1660645406953308170)


**Precision:**
Imagine you’re a detective investigating a series of crimes. You want to catch the right suspect without falsely accusing innocent people.
Precision is like your accuracy in identifying the actual criminals among the suspects you arrest.

Formula: Precision = (True Positives) / (True Positives + False Positives)

**Recall:**
Now, think about the same detective scenario. You want to make sure you don’t miss any criminals, even if it means arresting a few innocent people.
Recall measures how well you identify all the actual criminals, regardless of false arrests.
High recall means you catch most of the criminals, even if you occasionally arrest innocent folks.


Formula: Recall = (True Positives) / (True Positives + False Negatives)

### Decision Tree

image.png

Image Source: [Link](https://www.devops.ae/decision-tree-classification-algorithm/)

```python
# subset of predictors only
decision_tree_data = df_with_dummies.drop(['Churn_onehotenc', 'Churn_Yes'], axis=1)
```

```python
# split into train and test:
x_train,x_test,y_train,y_test = train_test_split(decision_tree_data,y_true,test_size=0.2)
```

```python
# Define the model parameters and fit the model
dec_tree = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=6, min_samples_leaf=8)
```

```python
dec_tree.fit(x_train,y_train)
```

```python
y_pred=dec_tree.predict(x_test)
y_pred
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

**NOTE:**
1. When looking at performance in case of an imbalanced classification problem, always look at the Precision and Recall of the smaller class.
2. Both Precision and Recall should be balanced (higher the better).
3. Accuracy is a good metric to look at when the class size is more or less equal (balanced dataset)

```python
# weighted decision tree
# weighting the classes - n_samples/(2*n_samples_class_A)
# Define the model parameters and fit the model
dec_tree_weighted = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=6, min_samples_leaf=8, class_weight='balanced')
```

```python
dec_tree_weighted.fit(x_train, y_train)
```

```python
y_pred=dec_tree_weighted.predict(x_test)
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
# ROC Curves
y_pred_proba = dec_tree_weighted.predict_proba(x_test)[:,1]
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import RocCurveDisplay

display = RocCurveDisplay.from_predictions(
    y_test,
    y_pred_proba,
    name="ROC Curve",
    color="darkorange",
    #plot_chance_level=True
)
display.ax_.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="ROC Curve: Churn vs No Churn",
)
```

```python
from sklearn.metrics import PrecisionRecallDisplay

display = PrecisionRecallDisplay.from_estimator(
    dec_tree_weighted, x_test, y_test, name="Weighted Decision Tree"
)
_ = display.ax_.set_title("2-class Precision-Recall curve")
```

```python
# let's choose a different threshold:
y_pred_proba = dec_tree_weighted.predict_proba(x_test)[:,1]
y_pred_035 = (y_pred_proba>=0.65).astype(int)
```

```python
print(classification_report(y_test, y_pred_035, labels=[0,1]))
```

### How to choose the right threshold?
- High threshold -> High precision, Low recall
- Looking at ROC/Precision-recall curves gives us a good idea
- What is more expensive for the business: false positive or false negative?
- Healthcare - false negatives are more expensive
